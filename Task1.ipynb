```python
import pandas as pd
```

```python
df = pd.read_csv("D:\\IVY PRO SCHOOL\\elevate\\22.09.2025ABI\\dataset 4\\netflix_titles_nov_2019.csv")
```

```python
df.head(3)
```

```python
## Identify and handle missing values using.
df.isnull().sum()
```

```python
##drop missing values
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
## Remove duplicate rows using.
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
##Standardize text values
df['title'] = df['title'].str.title()
df['director'] = df['director'].str.title()
df['cast'] = df['cast'].str.title()
df['country'] = df['country'].str.title()
df['rating'] = df['rating'].str.title()
df['listed_in']=df['listed_in'].str.title()
df['description'] = df['description'].str.title()
df['tyoe'] = df['type'].str.title()
```

```python
df.head()
```

```python
##rename column header
df.columns = [col.capitalize() for col in df.columns]
```

```python
df.head(3)
```

```python
## Convert date formats to a consistent type 
df["Date_added"] = pd.to_datetime(df["Date_added"], format="%B %d, %Y")
                                  
```
